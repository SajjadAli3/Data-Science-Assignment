# Add Column in Pandas

```python
import pandas as pd
```

```python
dataset=pd.read_csv("Pokemon.csv")
```

```python
dataset.head()
```

```python
# High_Low_HP & High_Low_Speed
HP_mean = dataset["HP"].mean()
Speed_mean = dataset["Speed"].mean()
```

```python
HP_mean
```

```python
Speed_mean
```

```python
def HP(val):
    if val >= HP_mean:
        return "High HP"
    else:
        return "Low HP"
def Speed(val):
    if val >= Speed_mean:
        return "High Speed"
    else:
        return "Low Speed"
    
```

```python
dataset["High_Low_HP"] = dataset["HP"].apply(HP)
```

```python
dataset["High_Low_Speed"] = dataset["Speed"].apply(Speed)
```
